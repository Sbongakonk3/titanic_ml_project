```python
import pandas as pd

df= pd.read_csv('../data/titanic.csv')
df.head()
```

```python
# Show how many missing values
df.isnull().sum()

# IMPUTATION is when we fill in missing values. 
# We use mean if data is symmetric without extreme outliers.
# If data is skewed we use MEDIAN IMPUTATION (e.g., more kids than adults).
# Modal imputation is more useful when we deal with categorical data.

# For "Age" maybe data was lost/not recorded. 
# We can't delete these people (we'd lose ~20% of our data), so we use Imputation.
# To fix this for Age, we replace with the middle value (Median Imputation).
# fillna searches for "NaN" values.

df['Age'] = df['Age'].fillna(df['Age'].median())

# DELETE (DROP) Cabin because it's noise.
# NB: inplace=True means that we change the DataFrame directly, not returning a copy.
df.drop(columns=['Cabin'], inplace=True)

# Show how many missing values remain after imputation and dropping Cabin
df.isnull().sum()

df.head()
```

```python
##IN OUR MODEL WE REPLACE WORDS WITH NUMBERS LIKE ASSIGN MALE TO 0 AND FEMALES TO 1
## WE CAN USE :

## y=w_1*SEX + w_2*AGE + b

# Create a dictionary to map the values
gender_map = {'male': 1, 'female': 0}

# Apply the map to the 'Sex' column
df['Sex'] = df['Sex'].map(gender_map)

# Check the table again
df.head()
```

```python
# List the columns we want to throw away because they don't help our model
cols_to_drop = ['Name', 'Ticket', 'PassengerId']

# Drop them from our DataFrame
df.drop(columns=cols_to_drop, inplace=True)

# Let's see our clean 'Matrix'
print(df.head())
```

```python
##LEARNING ONE-HOT ENCODING
import pandas as pd

# Example Titanic data with 'Embarked' column
data = {
    'PassengerId': [1, 2, 3, 4],
    'Embarked': ['S', 'C', 'Q', 'S']
}
df = pd.DataFrame(data)

# Show the original DataFrame
print("Original DataFrame:")
print(df)

#  One-Hot Encoding
# pd.get_dummies automatically converts categorical values into binary flags (0/1).
# columns=['Embarked'] → tells pandas which column to encode
# drop_first=True → drops the first category ('C') to avoid redundancy
# Why? Because if Embarked_S=0 and Embarked_Q=0, we already know the passenger embarked at C.
# Keeping all three columns would cause multicollinearity (perfectly predictable columns),
# which breaks some ML algorithms.
df_encoded = pd.get_dummies(df, columns=['Embarked'], drop_first=True)

# Show the transformed DataFrame
print("\nDataFrame after One-Hot Encoding:")
print(df_encoded)

# ✅ One-Hot Encoding in Action:
# Embarked_S → 1 if embarked at S, else 0
# Embarked_Q → 1 if embarked at Q, else 0
# If both are 0 → embarked at C (the dropped category)

# Final check of column data types
print("\nColumn data types:")
print(df_encoded.dtypes)
```

```python
# Convert 'Embarked' into multiple binary columns

## pd.get_dummies → built‑in Pandas function that converts categorical
## drop_first=True → drops the first category (C in this case) to avoid redundancy.
## Because if both Embarked_S=0 and Embarked_Q=0, the model automatically knows the passenger embarked at C.

df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)

# Final check of our data types
print(df.dtypes)
df.head()
```

```python
# The Target (y) is what we want to predict
# It’s the answer key: did the passenger survive (1) or not (0)?

y = df['Survived'] #This selects the column "Survived" from the Titanic dataset
#Shape: (891,) → a vector with 891 entries (one per passenger).

# The Features (X) are the clues we use to predict it
# We drop 'Survived' from X because the model shouldn't see the answer!

X = df.drop(columns=['Survived']) # i.e use every column except for "survive" to calculate "survive"
# We hide the answers from model

print("Target Shape (vector):", y.shape)
print("Feature Matrix Shape:", X.shape)
```
